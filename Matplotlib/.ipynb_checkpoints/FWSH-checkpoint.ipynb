```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Apply Seaborn style for a modern look
sns.set_style("whitegrid")
fifa = pd.read_csv('fifa_data.csv')
# Define bins
bins = [40, 50, 60, 70, 80, 90, 100]

# Create figure
plt.figure(figsize=(9, 5))

# Plot histogram with better color styling
plt.hist(fifa.Overall, bins=bins, color=plt.get_cmap("Blues")(0.6), edgecolor="black", linewidth=1.2)

# Set x-ticks for better readability
plt.xticks(bins, fontsize=12, fontweight="bold", color="black")

# Customize labels
plt.ylabel("Number of Players", fontdict={'fontname': 'Myanmar Text', 'fontsize': 14, 'color': 'black', 'fontweight': 'bold'})
plt.xlabel("Skill Level", fontdict={'fontname': 'Myanmar Text', 'fontsize': 14, 'color': 'black', 'fontweight': 'bold'})

# Title customization
plt.title("ðŸ“Š FIFA 2018 Player Skill Distribution", fontsize=18, color='darkblue', fontweight='bold', pad=15)

# Add a subtle grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
